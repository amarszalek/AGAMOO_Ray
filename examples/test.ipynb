```python
import ray
from agamoo_ray.players import ClonalSelection
from agamoo_ray.objectives import RE31, RE32, RE33, RE34, RE35, RE36, RE37
from agamoo_ray import AGAMOO, Evaluator
import logging
import os

os.environ["RAY_ACCEL_ENV_VAR_OVERRIDE_ON_ZERO"] = "0"
logging.basicConfig(level=logging.INFO)
```

```python
max_eval = 100000
npop = 25 
change_iter = 1
next_iter = -1
max_front = 200
player_parm = { "nclone": 12, "mutate_args": [0.45, 0.9, 0.01], 'sup': 0.0}
```

```python
obj1 = RE31(0, obj=1)
obj2 = RE31(1, obj=2)
obj3 = RE31(2, obj=3)

objs = [obj1, obj2, obj3]

nvar = obj1.n_var
nobj = obj1.n_obj
```

```python
ray.init(log_to_driver=True)

agamoo = AGAMOO(max_eval, change_iter, next_iter, max_front)
storage = agamoo.create_storage(nvar, nobj)
evaluator = Evaluator.remote(objs)
players = [ClonalSelection.remote(i, npop, player_parm, obj, storage, exchange='front_random') for i, obj in zip(range(nobj), objs)]
agamoo.init_players(players, evaluator)
```

```python
agamoo.start_optimize()
```
